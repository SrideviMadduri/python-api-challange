```python
import gmaps
import pandas as pd

# Google developer API key
from api_keys import gkey
```

```python
print(gkey)
```

```python
cities_df = pd.read_csv('output_data/cities.csv', index_col=0)
cities_df
```

```python
# Configure gmaps with API key
gmaps.configure(api_key=gkey)
```

```python
 #Store 'Lat' and 'Lng' into  locations 
locations = cities_df[["Lattitude", "Longitude"]].astype(float)

# Convert Poverty Rate to float and store
# HINT: be sure to handle NaN values
#cities_df = cities_df.dropna()
humidity = cities_df["Humidity"].astype(float)
```

```python
# Create a poverty Heatmap layer

fig = gmaps.figure(center=(0,0), zoom_level=2)
#gmaps.figure()

heat_layer = gmaps.heatmap_layer(locations, 
                                 weights=humidity, 
                                 dissipating=False, max_intensity=100,
                                 point_radius = 5
                                 
                                )

fig.add_layer(heat_layer)

fig
```

```python
#Narrow down the DataFrame to find your ideal weather condition. For example:
#  * A max temperature lower than 80 degrees but higher than 70.
#  * Wind speed less than 10 mph.
 # * Zero cloudiness.


ideal_weather_df =cities_df[(cities_df['Temperature'] > 60) & (cities_df['Temperature']<80)
                            & ((cities_df['Wind Speed']<10)) & (cities_df['Cloudiness']==0)] 

ideal_weather_df
```

```python
bank_layer = gmaps.symbol_layer(
    locations, fill_color='rgba(0, 150, 0, 0.4)',
    stroke_color='rgba(0, 0, 150, 0.4)', scale=2,
    info_box_content=[f"<h3>Bank amount:</h3> {bank}</br>Sudeep" for bank in range(locations.shape[0])]
    #info_box_content=['a','b']
    #hover_text=['x' for x in range(locations.shape[0])]
)


fig = gmaps.figure()
fig.add_layer(bank_layer)

fig
```
